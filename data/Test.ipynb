```python
from pathlib import Path

import geopandas as gpd
import osmnx as ox

ox.__version__

ox.settings.use_cache = True

ox.settings.log_console = True

bbox = 12.08625, 55.73939, 12.14461, 55.78072

G = ox.graph.graph_from_bbox(bbox, simplify = False, network_type="drive_service") 
#city_proj = ox.projection.project_gdf(city)

fig, ax = ox.plot.plot_figure_ground(G, dist = 4000)
_ = ax.axis("off")
```

```python
places = [
    "Copenhagen Municipality, Denmark",
    "Frederiksberg Municipality, Denmark",
]

tags = {
    "highway": True,              
    "leisure": True,            
    "amenity": ["school", "university", "college"], 
    "public_transport": True,     
}


features = ox.features.features_from_place(places, tags)

ax = features.plot(
    column=features.geom_type,
    figsize=(10, 10),
    legend=True
)

G = ox.graph.graph_from_place(places, simplify = True, network_type="walk")
fig, ax = ox.plot.plot_graph(G, node_size=0, edge_color="w", edge_linewidth=0.2)

features.plot(
    ax=ax,
    alpha=0.7)

ax.axis("off")

```
